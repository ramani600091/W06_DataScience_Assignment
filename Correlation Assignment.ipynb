```python
# ===============================================
# 📊 Correlation Analysis on PrePlacement Dataset
# ===============================================

# 1️⃣ Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2️⃣ Load Dataset
dataset = pd.read_csv("PrePlacement.csv")

# Select only relevant columns
dataset = dataset[['sl_no', 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']]

print("✅ Dataset Loaded Successfully!\n")
print(dataset.head())

# 3️⃣ Check for Missing Values
print("\nMissing values per column:")
print(dataset.isnull().sum())

# 4️⃣ Generate Pairplot for Overall Numeric Relationships
sns.pairplot(dataset, diag_kind='kde')
plt.suptitle("Pairwise Relationships Between All Numeric Variables", y=1.02)
plt.show()

# 5️⃣ Pairplot Focused on Salary and Performance Features
selected_cols = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']
sns.pairplot(dataset[selected_cols], diag_kind='kde')
plt.suptitle("Pairplot Focusing on Salary and Related Performance Metrics", y=1.02)
plt.show()

# 6️⃣ Compute Correlation Matrix
corr_matrix = dataset.corr(numeric_only=True)
print("\n📈 Correlation Matrix:")
print(corr_matrix)

# 7️⃣ Visualize Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of PrePlacement Dataset")
plt.show()

# 8️⃣ Identify Feature Most Correlated with Salary
salary_corr = corr_matrix['salary'].drop('salary').sort_values(ascending=False)
most_corr_feature = salary_corr.index[0]
print("\n🔹 Correlation of Each Feature with Salary:")
print(salary_corr)

print(f"\n✅ The parameter most highly correlated with salary is **{most_corr_feature}**, "
      f"with a correlation coefficient of {salary_corr[most_corr_feature]:.3f}.")
```
